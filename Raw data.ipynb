```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
import numpy as np
```

```python
urls = pd.read_csv('urls.csv')
urls.sort_values(['Year', 'Model'], ascending=[False, True], inplace=True)
urls.reset_index(drop=True, inplace=True)
urls.to_csv('urls.csv', index=False)
```

```python
import requests
def scarpe_page(url):
    request_object = requests.get(url)

    soup = BeautifulSoup(request_object.text, 'html.parser')
    grid = soup.findAll('table', ({'class': 'Grid'}))
    if(len(grid) == 0): return
        
    location = -1
    for i in range(len(grid)):
        grid_title = grid[i].findAll('tr')[0].text
        if 'General' in grid_title and 'information' in grid_title: location = i
        
    all_tr = grid[location].findAll('tr')
    entry = {}

    for tr in all_tr:
        content = tr.findAll('td')
        if content != [] and len(content) == 2:
            entry[content[0].text.split(':')[0]] = content[1].text
    
    return entry
```

```python
all_entries = {}

for i in range(urls.shape[0]):
    if i%50 == 0: print(i)
    all_entries[i] = scarpe_page(urls.iloc[i]['URL'])
```

```python
for key, value in list(all_entries.items()):
    if value == None: del all_entries[key]
```

```python
import json
with open('raw.json', 'w') as fp:
    json.dump(all_entries, fp)
```
